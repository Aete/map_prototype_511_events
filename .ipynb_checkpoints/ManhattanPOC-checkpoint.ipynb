## Will use the generated file to combine with crash data geojson, Now extract NYC Collision data

```python
# I have downloaded nyc_collision data in current directory
nyc_collision = pd.read_csv("nyc_collisions.csv")
```

```python
nyc_collision.head(10)
```

```python
# Getting crash data for Manhattan Borough and Q4 2019
nyc_collision.info()
```

```python
start_date = '01-10-2019'
end_date = '31-12-2019'
nyc_collisionQ4Manhattan = nyc_collision[(pd.to_datetime(nyc_collision['CRASH DATE']) >= start_date) & (pd.to_datetime(nyc_collision['CRASH DATE']) <= end_date) & (nyc_collision['BOROUGH'] == 'MANHATTAN') ]
```

```python
nyc_collisionQ4Manhattan.head()
```

```python
# for missing point location we can match with OSM id using street name if that matches
# for now I am removing NaN under Location
nyc_collisionQ4Manhattan.shape
```

```python
nyc_collisionQ4Manhattan['LOCATION'].isnull().sum()
```

```python
nyc_collisionQ4ManhattanClean = nyc_collisionQ4Manhattan.dropna(axis=0, subset=['LOCATION'])
```

```python
nyc_collisionQ4ManhattanClean.head()
```

```python
nyc_collisionQ4ManhattanClean.shape
```

```python
nyc_collisionQ4ManhattanClean.rename(columns={'ZIP CODE': 'ZIP_CODE'}, inplace=True)
```

```python
nyc_collisionQ4ManhattanCleanSubset = nyc_collisionQ4ManhattanClean[['ZIP_CODE', 'LATITUDE', 'LONGITUDE']]
```

```python
# Saving the clean Crash data to csv
nyc_collisionQ4ManhattanCleanSubset.to_csv("manhattan_Q4_2019_coliision.csv", index= False)
```

```python
nyc_collisionQ4ManhattanCleanSubset.head()
```

```python
nyc_man_coll_gdf = gp.GeoDataFrame(
    nyc_collisionQ4ManhattanCleanSubset, geometry=gp.points_from_xy(nyc_collisionQ4ManhattanCleanSubset.LONGITUDE, 
                                                                    nyc_collisionQ4ManhattanCleanSubset.LATITUDE))
```

```python
type(nyc_man_coll_gdf)
```

```python
nyc_man_coll_gdf.plot()
# this data contains crash point with position as (0,0), removing those for instead of geocoding them
```

```python
nyc_man_coll_gdf_clean = nyc_man_coll_gdf[nyc_man_coll_gdf.LATITUDE != 0]
nyc_man_coll_gdf_clean.plot()
```

```python
# Saving this to file to process with sharedstreets tool
nyc_man_coll_gdf_clean.to_file("ManhattanCrashQ42019.geojson", driver="GeoJSON")
# after processing the file I got 3 files as output: matched, unmatched, clustered geojson, Lets read clustered one
```
